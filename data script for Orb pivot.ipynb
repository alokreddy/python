```python
def header_filler(df):
    df.columns = [c.replace(' ', '_') for c in df.columns]

header_filler(hr)
header_filler(er)
```

```python
df = er.merge(hr, on='PID', how='left')
df = df[df.Assignment_Category_x.isin(['Regular', 'Regular Limited']) & df.Career_Level.notnull()]
```

```python
frame_dict = {'VP Criteria' : ["EX5","EX4","EX3","EX2","EX1"]
           ,'Management' : ["M4", "M3", "M2", "M1"]
           , 'Professional' : ["P6", "P5", "P4", "P3", "P2", "P1"]
           , 'Sales' : ["S6", "S5", "S4", "S3", "S2", "S1"]
           , 'Support' : ["U4", "U3", "U2", "U1"]
           , 'Scientific' : ["T6","T5","T4", "T3", "T2"]
           , 'Production/Operations' : ["W4", "W3", "W2", "W1"]}

joblevel_dict = {
    'EX5' : 'CEO',
    'EX4' : 'CEO Directs',
    'EX3' : 'BRC',
    'EX2' : 'Corporate Vice Presidents',
    'EX1' : 'Vice President',
    'M4' : 'Senior Director',
    'M3' : 'Associate Director & Director',
    'M2' : 'Manager & Senior Manager',
    'M1' : 'Associate Manager',
    'T6' : 'Senior Master',
    'T5' : 'Master',
    'T4' : 'Expert',
    'T3' : 'Senior / Career Level',
    'T2' : 'Intermediate',
    'P5' : 'Master',
    'P4' : 'Expert',
    'P3' : 'Senior / Career Level',
    'P2' : 'Intermediate',
    'P1' : 'Entry',
    'S5' : 'Sales Director',
    'S4' : 'Expert',
    'S3' : 'Senior / Career Level',
    'S2' : 'Intermediate',
    'S1' : 'Entry',
    'U3' : 'Senior',
    'U2' : 'Intermediate',
    'U1' : 'Entry',
}
```

```python
for k, v in frame_dict.iteritems():
    df.loc[df["Career_Level"].isin(v), "Framework"] = k


for k, v in joblevel_dict.iteritems():
    df.loc[df["Career_Level"].eq(k), "Job Level"] = v

df = df[df['Job Level'].notnull()]
```

```python
df = df[['Career_Band', 'Framework','Career_Level', 'Job Level', 'Job_Title_Description_x', 'SBU_x', 'Service_Line_x', 'Cost_Center_x', 'BRC', 'BRC-1', 'BRC-2']].drop_duplicates()
#df = df[~df['Career_Level'].isin(['P6', 'S6'])]
```

```python
df.columns = [c.replace('_x', '') for c in df.columns]
df.columns = [c.replace('_', ' ') for c in df.columns]
```

```python
from itertools import izip
cl_list = ["EX5", "EX4", "EX3", "EX2", "EX1", "M4", "M3", "M2", "M1", "T6","T5","T4", "T3", "T2","P5", "P4", "P3", "P2", "P1", "S5", "S4", "S3", "S2", "S1", "U4", "U3", "U2", "U1", "W4", "W3", "W2", "W1"]
custom_dict = dict(izip(cl_list,range(len(cl_list))))
df['rank'] = df['Career Level'].map(custom_dict)
df.sort_values(by=['rank', 'Job Title Description'],ascending=[True, True], inplace=True)

df.drop(['Career Band', 'rank'], axis=1, inplace=True)

from pprint import pprint as pp
import re
stringy = df.to_json(orient="values")
re.sub('\\\\','',stringy)
```
